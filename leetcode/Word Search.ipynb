```python
word = 'ABCCED'
board = [
  ['A','B','C','E'],
  ['S','F','C','S'],
  ['A','D','E','E']
]

# a little broken
def find_iter(word, board):
    # find possible starting points
    starts = [(x, y) for x, row in enumerate(board) 
                     for y, char in enumerate(row)
                     if char == word[0]]
    print(starts)
    
    for start in starts:
        neighbours = [start]
        levels = [0]
        visited = [[False] * len(row) for row in board]
        curr_word = word
        prev_level = 0
        
        while len(curr_word) > 0 and len(neighbours) > 0:
            print(curr_word, neighbours)
            x, y = neighbours.pop()
            level = levels.pop()
            
            # if searched all of this level, undo last char
            if level < prev_level:
                curr_word = word[len(word) - len(curr_word) -1:]
                # undo visited
                print('undo to', curr_word)
            

            # if havent used char, and char matches
            if not visited[x][y] and board[x][y] == curr_word[0]:
                # mark as visited
                visited[x][y] = True
                
                # clear neighbours
                # neighbours = []

                # remove char from word
                curr_word = curr_word[1:]

                # add its neighbours
                if x > 0: 
                    neighbours.append((x-1, y))
                    levels.append(level+1)
                if x < len(board) - 1: 
                    neighbours.append((x + 1, y))
                    levels.append(level+1)
                if y > 0: 
                    neighbours.append((x, y - 1))
                    levels.append(level+1)
                if y < len(board[0]) - 1: 
                    neighbours.append((x, y + 1))
                    levels.append(level+1)

            prev_level = level
            
            
        if len(curr_word) == 0:
            return True
    
    return False
    
```

```python
from copy import deepcopy

def copy_used(used): 
    return [[cell for cell in row] for row in used]
        
# recursive
def find(word, board):
    used = [[False] * len(row) for row in board]
    starts = [(x, y) for x, row in enumerate(board) 
                     for y, char in enumerate(row)
                     if char == word[0]]
    for start in starts:
        print('start', start)
        if find_rec(word, [start], board, used):
            return True
        print()
    return False
    
def find_rec(word, candidates, board, used):
    print(word, candidates)
    # base case
    if len(word) <= 1:
        for x,y in candidates:
            if not used[x][y] and board[x][y] == word[0]:
                return True
        return False
    
    # if word[0] in candidates, and not visited,
    for x,y in candidates:
        if not used[x][y] and board[x][y] == word[0]:
            used [x][y] = True
            left = right = top = bottom = False
            # find if rest of word is in neighbours
            if x > 0: 
                top = find_rec(word[1:], [(x-1, y)], board, used)
            if x < len(board) - 1: 
                bottom = find_rec(word[1:], [(x+1, y)], board, used)
            if y > 0: 
                left = find_rec(word[1:], [(x, y-1)], board, used)
            if y < len(board[0]) - 1: 
                right = find_rec(word[1:], [(x, y+1)], board, used) 
            
            found_any = left or right or top or bottom
            
            if not found_any:
                used[x][y] = False
                
            return found_any
    
    # else return false
    return False
```
