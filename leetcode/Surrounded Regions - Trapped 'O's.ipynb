# Surrounded Regions - Trapped 'O's
[link](https://leetcode.com/problems/surrounded-regions/)

Given a 2D board containing 'X' and 'O', capture all regions surrounded by 'X'.

A region is captured by flipping all 'O's into 'X's in that surrounded region.

For example,

    X X X X
    X O O X
    X X O X
    X O X X
    
After running your function, the board should be:

    X X X X
    X X X X
    X X X X
    X O X X


```python
def trapped(board, coord=(0,0), visited=None):
    # empty board
    if len(board) == 0:
        return False
    
    # initialize visited tracker
    if visited is None:
        visited = [[False] * len(row) for row in board]
    
    # mark cell as visited
    x, y = coord
    visited[x][y] = True
    
    # valid neighbouring coordinates
    ns = [(i, j) for i, j in [(x+1, y), (x-1, y), (x, y+1), (x,y-1)]
                        if i >= 0 and  i < len(board) and
                            j >= 0 and j < len(board[0])]
    
    # if is O, set to X if neighbouring Os are also trapped
    if board[x][y] == 'O':
        neighbours_trapped = True
        for n in ns: 
            i, j = n
            # if there is neighbouring O, check if also trapped
            if board[i][j] == 'O' and not visited[i][j]:
                if not trapped(board, n, visited):
                    neighbours_trapped = False
                    break
        # if neigbouring is trapped on all 4 sides (no escape on edge)
        if neighbours_trapped and len(ns) == 4:
            board[x][y] = 'X'
            return True
            
    # if isn't O, search N,S,E, and W for O
    else:
        for n in ns:
            i, j = n
            if not visited[i][j]:
                trapped(board, n, visited)

    return False
```
